```python
from extractFeatures import *
import librosa
import os
import numpy as np
import pandas as pd
from pydub import AudioSegment
```

```python
input_dir= "data/full_length/mp3"
output_wav = "data/full_length/wav"   # Your new wav dataset
data = [] # List to store feature dictionaries
```

```python
# Convert from mp3 to wav
for label in os.listdir(input_dir):
    label_dir = os.path.join(input_dir, label)

    if os.path.isdir(label_dir):
        print(f"Processing folder: {label}")

        output_label_dir = os.path.join(output_wav, label)
        os.makedirs(output_label_dir, exist_ok=True)

        for file in os.listdir(label_dir):
            if file.endswith(".mp3"):
                mp3_path = os.path.join(label_dir, file)
                wav_filename = os.path.splitext(file)[0] + ".wav"
                wav_path = os.path.join(output_label_dir, wav_filename)

                # Convert mp3 to wav
                audio = AudioSegment.from_mp3(mp3_path)
                audio.export(wav_path, format="wav")
                print(f"  Converted: {file} -> {wav_filename}")
```

```python
# Iterate through the dataset and extract features
input_dir = "data/full_length/wav"

for label in os.listdir(input_dir):
    label_dir = os.path.join(input_dir, label)

    if os.path.isdir(label_dir):  # Ensure it's a folder
        print(f"Processing folder: {label}")  # Print the folder being processed
        
        for file in os.listdir(label_dir):
            if file.endswith(".wav"):
                file_path = os.path.join(label_dir, file)
                print(f"  Extracting features from: {file}")  # Print the file name being processed
                
                # Load audio file
                y_audio, sr = librosa.load(file_path, sr=22050)

                # Extract features
                features = extract_features(y_audio, sr)
                
                # Add filename and label
                features["file"] = file
                features["label"] = label  # Assign label from folder name

                # Append to list
                data.append(features)

# Convert to DataFrame
df = pd.DataFrame(data)
```

```python
# Save to CSV
output_csv = "data/features_full_v7.csv"
df.to_csv(output_csv, index=False)
print(f"Features saved to {output_csv}")
```
